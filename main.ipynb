Project Title : 
Sales Performance Analysis & Visualization

There is amazon sales data which has multiple user's sales data so on this data we get analysis and predication which get further information.


Import Amazon Sales Data

```python
import pandas as pd

df = pd.read_csv("amazon_sales.csv")
df.head()
```

Checking Null Data and Data Cleaning 

```python
df.isnull().sum()

# Remove rows with missing values
df = df.dropna()
df.info()

# Clean Price column
df['discounted_price'] = df['discounted_price'].replace('[₹,]', '', regex=True).astype(float)
df['actual_price'] = df['actual_price'].replace('[₹,]', '', regex=True).astype(float)

# Clean Discount Percentage column
df['discount_percentage'] = df['discount_percentage'].replace('%', '', regex=True).astype(float)
df.head()
```

```python
# Convert Rating & Count
df['rating'] = pd.to_numeric(df['rating'], errors='coerce')
df['rating_count'] = df['rating_count'].replace(',', '', regex=True)
df['rating_count'] = pd.to_numeric(df['rating_count'], errors='coerce')

# Clean Category column
df['main_category'] = df['category'].apply(lambda x: x.split('|')[0])
df.head()
```

```python
# New Column Discount Amount
df['discount_amount'] = df['actual_price'] - df['discounted_price']
```

```python
# Save cleaned data
df.to_csv("cleaned_amazon.csv", index=False)
```
